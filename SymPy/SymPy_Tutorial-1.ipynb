```
import sympy as sym
import numpy as np 
import math
import matplotlib.pyplot as plt
```

```
x = sym.symbols('x') #   (x)
```

# Functions & Plotting

```
f = sym.sin(x)
g = sym.cos(x)
```

```
fx = sym.lambdify(x, f, 'numpy')
gx = sym.lambdify(x, g, 'numpy')
```

```
xVals = np.linspace(-math.pi,math.pi,100)
```

```
plt.plot(xVals, fx(xVals), label = 'f(x) = sin(x)')
plt.plot(xVals, gx(xVals), label = 'g(x) = cos(x)')
plt.xlabel('x')
plt.ylabel('f(x) = sin (x)')
plt.title('f(x) = sin (x)')
plt.show()
```

# Inverse Plots

```
f = sym.asin(x)
g = sym.acos(x)
```

```
fx = sym.lambdify(x, f, 'numpy')
gx = sym.lambdify(x, g, 'numpy')
```

```
xVals = np.linspace(-2,2,100)
plt.plot(xVals, fx(xVals), label = 'f(x) = asin(x)')
plt.plot(xVals, gx(xVals), label = 'g(x) = acos(x)')
plt.xlabel('x')
plt.ylabel('f(x), g(x)')
plt.title('Inverse Sines and Cosines')
plt.legend()
plt.show()
```

# Inverse Hyperbolic Sines an Cosines

```
f = sym.asinh(x)
g = sym.acosh(x)
```

```
fx = sym.lambdify(x, f, 'numpy')
gx = sym.lambdify(x, g, 'numpy')
```

```
xVals = np.linspace(1,10,100)
plt.plot(xVals, fx(xVals), label = 'f(x) = asinh(x)')
plt.plot(xVals, gx(xVals), label = 'g(x) = acosh(x)')
plt.xlabel('x')
plt.ylabel('f(x), g(x)')
plt.title('Inverse Hyperbolic Sines and Cosines')
plt.legend()
plt.show()
```

$$ f(x) = [x]$$
$$ g(x) = [x]$$
$$ h(x) = x - [x]$$

```
f = sym.ceiling(x) # return the smallest integer value which is not less than x
g = sym.floor(x) # return the largest integer value which is not greater than x
h = sym.frac(x) # return the fractional part of x
```

```
fx = sym.lambdify(x, f, 'numpy')
gx = sym.lambdify(x, g, 'numpy')
hx = sym.lambdify(x, h, 'numpy')
```

```
xVals = np.linspace(-10,10,100)
plt.plot(xVals, fx(xVals), label = 'f(x) = ceiling(x)')
plt.plot(xVals, gx(xVals), label = 'g(x) = floor(x)')
# plt.plot(xVals, hx(xVals), label = 'h(x) = frac(x)')  # Showing error 
plt.xlabel('x')
plt.ylabel('f(x), g(x)')
plt.title('Hyperbolic Sines and Cosines')
plt.legend()
plt.show()
```

$$f(x) = e^x$$
$$g(x) = log(x)$$

```
f = sym.exp(x)
g = sym.log(x)
```

```
fx = sym.lambdify(x, f, 'numpy')
gx = sym.lambdify(x, g, 'numpy')
```

```
xVals = np.linspace(0.01,1,100)
plt.plot(xVals, fx(xVals), label = 'f(x) = exp(x)')
plt.plot(xVals, gx(xVals), label = 'g(x) = log(x)')
plt.xlabel('x')
plt.ylabel('f(x), g(x)')
plt.title('Exponents and Logs')
plt.legend()
plt.show()
```
